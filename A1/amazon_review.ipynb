Import of the neccesary libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
folder = ""
reviews_learn_dataset_path = folder + "amazon_review_ID.shuf.lrn.csv"
reviews_test_dataset_path = folder + "amazon_review_ID.shuf.tes.csv"
```

Reading in the CSV-file.

```python
rev_df_lrn_original = pd.read_csv(reviews_learn_dataset_path)
rev_df_tes_original = pd.read_csv(reviews_test_dataset_path)

rev_df_lrn = rev_df_lrn_original.copy()
rev_df_tes = rev_df_tes_original.copy()

print("Train dataset")
print(rev_df_lrn.head())

print("Test dataset")
print(rev_df_tes.head())
```

Some basic information about the dataset

```python
def dataset_info(df, split_type):
  shape = df.shape
  print("The dataset consists of: " + str(shape[0]) +  " rows and " + str(shape[1]) + " columns")
  if split_type == "train":
    unique_values = df['Class'].unique()
    print("The dataset has " + str(len(unique_values)) + " unique values in the target attribute")
    print("The values are: ")
    print(unique_values)
  print("\n")

  na_col = df.isna().sum()
  if na_col[na_col > 0].shape == (0,):
      print("There are no missing values in the dataset")
  else:
    print("The columns with NA values are: ")
    na_col[na_col > 0]

  print("\n")
  d_types = pd.Series(df.dtypes)
  print("The datatypes in the frame are:")
  print(d_types.value_counts())
```

```python
print("Training dataset information: ")
print("\n")
dataset_info(rev_df_lrn, "train")
print("\n")
print("Test dataset information: ")
print("\n")
dataset_info(rev_df_tes, "test")
```

To be able to work with target attributes in a classification task, we need a numeric representation of the different classes.
Therefore we encode the different text categories with a numeric representation.

It isn't strictly neccesary to encode the classes in the target attribute numericaly, in terms of the theory beind classification algorithms, but most frameworks require a numerical representation since it makes comparisons and storage more efficient.

To do this, SKLearns preprocessing library is used, and the orignal categorires are stored in a variable

```python
le = LabelEncoder()

rev_df_lrn['Class'] = rev_df_lrn['Class'].astype('category')
rev_df_lrn_text_categories = rev_df_lrn['Class'].astype('category')
rev_df_lrn['Class_encoded'] = le.fit_transform(rev_df_lrn['Class'])
rev_df_lrn['Class_encoded'] = rev_df_lrn['Class_encoded'].astype('category')

le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("The characters encodings are: ")
print(le_name_mapping)

print("Dataframe after class encoding: ")
rev_df_lrn = rev_df_lrn.drop('Class', axis=1)

rev_df_lrn.head()
```

Some model really on meassuring the numeric distance between samples, and there the value range should not have an impact. To not let the value range have an impact, the data can be scaled.

There are several ways to do this, here the values will be adjusted using min-max scaling and the z-score method.  

Min max scaling:

```python
rev_df_lrn_minmax_scaled = rev_df_lrn.copy()
rev_df_tes_minmax_scaled = rev_df_tes.copy()


for column in rev_df_lrn_minmax_scaled.iloc[:, 1:10001].columns:
    rev_df_lrn_minmax_scaled[column] = (rev_df_lrn_minmax_scaled[column] - rev_df_lrn_minmax_scaled[column].min()) / (rev_df_lrn_minmax_scaled[column].max() - rev_df_lrn_minmax_scaled[column].min())

for column in rev_df_tes_minmax_scaled.iloc[:, 1:10001].columns:
    rev_df_tes_minmax_scaled[column] = (rev_df_tes_minmax_scaled[column] - rev_df_tes_minmax_scaled[column].min()) / (rev_df_tes_minmax_scaled[column].max() - rev_df_tes_minmax_scaled[column].min())



rev_df_lrn_minmax_scaled.head()
```

Z-score standardisation:

```python
rev_df_lrn_z_scaled = rev_df_lrn.copy()
rev_df_tes_z_scaled = rev_df_tes.copy()


for column in rev_df_lrn_z_scaled.iloc[:, 1:10001].columns:
    rev_df_lrn_z_scaled[column] = (rev_df_lrn_z_scaled[column] - rev_df_lrn_z_scaled[column].mean()) / rev_df_lrn_z_scaled[column].std()

for column in rev_df_tes_z_scaled.iloc[:, 1:10001].columns:
    rev_df_tes_z_scaled[column] = (rev_df_tes_z_scaled[column] - rev_df_tes_z_scaled[column].mean()) / rev_df_tes_z_scaled[column].std()


rev_df_lrn_z_scaled.head()
```

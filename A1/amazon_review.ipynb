Some basic information about the dataset

```python
def dataset_info(df, split_type):
  shape = df.shape
  print("The dataset consists of: " + str(shape[0]) +  " rows and " + str(shape[1]) + " columns")
  if split_type == "train":
    unique_values = df['Class'].unique()
    print("The dataset has " + str(len(unique_values)) + " unique values in the target attribute")
    print("The values are: ")
    print(unique_values)
  print("\n")

  na_col = df.isna().sum()
  if na_col[na_col > 0].shape == (0,):
      print("There are no missing values in the dataset")
  else:
    print("The columns with NA values are: ")
    na_col[na_col > 0]

  print("\n")
  d_types = pd.Series(df.dtypes)
  print("The datatypes in the frame are:")
  print(d_types.value_counts())
```

```python
print("Training dataset information: ")
print("\n")
dataset_info(rev_df_lrn, "train")
print("\n")
print("Test dataset information: ")
print("\n")
dataset_info(rev_df_tes, "test")
```
